```python
from datetime import datetime, timedelta
import requests
import csv
import concurrent.futures

def get_dates(start_date, end_date):
    current = start_date
    while current <= end_date:
        yield current.strftime('%Y-%m-%d')
        current += timedelta(days=1)

def scrape_data(date):
    url = f"https://www.dailymetalprice.com/ajax-multi.php?d={date}&x=USD&k=0975cb203e4dae6017f3347d3d389db20db0f2fd69ee75c14659be16a404a093"
    
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36",
    "Cookie": ""
    }

    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            json_data = response.json()
            row = {"Date": json_data["date"]}
            
            for item in json_data["data"]:
                metal, price, _, unit, currency, _ = item
                column_name = f"{metal} ({currency}/{unit})"
                row[column_name] = price
            
            return row
        print(f"Failed to fetch {date}: Status {response.status_code}")
    except Exception as e:
        print(f"Error fetching {date}: {str(e)}")
    return None

start_date = datetime(2000, 1, 3)
end_date = datetime.today()

dates = list(get_dates(start_date, end_date))

all_data = []
with concurrent.futures.ThreadPoolExecutor(max_workers=20) as executor:
    future_to_date = {executor.submit(scrape_data, date): date for date in dates}
    for future in concurrent.futures.as_completed(future_to_date):
        date = future_to_date[future]
        try:
            result = future.result()
            if result:
                all_data.append(result)
        except Exception as e:
            print(f"Exception processing {date}: {e}")

all_data.sort(key=lambda x: x['Date'])

column_headers = ["Date"]
seen_headers = set(column_headers)
for row in all_data:
    for key in row:
        if key != 'Date' and key not in seen_headers:
            seen_headers.add(key)
            column_headers.append(key)

output_file = r"E:\Upwork\Australia_GDP\Commodities\metal_prices.csv"
with open(output_file, "w", newline="", encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=column_headers)
    writer.writeheader()
    writer.writerows(all_data)

print(f"Data saved to {output_file}")
```
